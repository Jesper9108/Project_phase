# Average, median and mode of file size

```python
def check_size(size):
    size = str(size).lower()
    if size == 'Varies with device':
        return np.nan
    elif size[-1] == 'm':
        size_float = float(size[:-1])
        return size_float
    elif size[-1] == 'k':
        size_float = float(size[:-1]) / 1024
        return size_float
    
modified_size = playstore.Size.apply(check_size)
```

```python
modified_size.agg(['mean', 'median'])
```

```python
 modified_size.mode()
```
