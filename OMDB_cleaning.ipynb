```python
import numpy as np
import pandas as pd
```

```python
omdb_initframe = pd.read_csv('E:/greenfox/ProjectPhase/third_week/omdb_datasets/omdb_0to100k.csv', sep=';', encoding='latin-1', low_memory=False)
```

```python
omdb_initframe.drop(columns=['Unnamed: 0'], inplace = True)
omdb_initframe.set_index(['imdbID'])
```

```python
omdb_initframe['Runtime'] = omdb_initframe['Runtime'].str.extract(r'(\d+)')
omdb_initframe['Runtime'] = omdb_initframe['Runtime'].astype('float64')
```

```python
omdb_initframe['Director'].replace({np.nan}, {''}, inplace=True)
```

# Removing commas from imdbVotes column 

```python
omdb_initframe['imdbVotes'].replace(r',', r'', regex=True, inplace = True)
omdb_initframe['imdbVotes']=omdb_initframe['imdbVotes'].astype('float64')
```

# Remove non-digit characters from Boxoffice column

```python
omdb_initframe['BoxOffice'] = omdb_initframe['BoxOffice'].str[1:]
```

```python
omdb_initframe.at[75940, 'BoxOffice'] = 25700
```

```python
omdb_initframe['BoxOffice'].replace(r',' , r'', regex=True, inplace = True)
omdb_initframe['BoxOffice'] = omdb_initframe['BoxOffice'].astype('float64')
```

# Convert imdb_Rating values between 0-1 format

```python
omdb_initframe['imdb_Rating'] = omdb_initframe['imdb_Rating'].str.extract(r'(^\d?\.\d)')
omdb_initframe['imdb_Rating'] = omdb_initframe['imdb_Rating'].astype('float64')
omdb_initframe['imdb_Rating'] = omdb_initframe['imdb_Rating']/10
```

# Convert RottenTomatoes_Rating values between 0-1 format

```python
omdb_initframe['RottenTomatoes_Rating'] = omdb_initframe['RottenTomatoes_Rating'].str.extract(r'(^\d+)')
omdb_initframe['RottenTomatoes_Rating'] = omdb_initframe['RottenTomatoes_Rating'].astype('float64')
omdb_initframe['RottenTomatoes_Rating'] = omdb_initframe['RottenTomatoes_Rating']/100
```

# Convert Metacritic_Rating values between 0-1 format

```python
omdb_initframe['Metacritic_Rating'] = omdb_initframe['Metacritic_Rating'].str.extract(r'(^\d+)')
omdb_initframe['Metacritic_Rating'] = omdb_initframe['Metacritic_Rating'].astype('float64')
omdb_initframe['Metacritic_Rating'] = omdb_initframe['Metacritic_Rating']/100
```

```python
omdb_initframe['Website'].replace(np.nan, '', inplace = True)
```

```python
Oscar_winner = omdb_initframe['Awards'].str.extract(r'(Won \d+ O\w+)', expand=False)
```

```python
Oscar_nominated = omdb_initframe['Awards'].str.extract(r'(^\w+ \w+ \d+ [O|G]\w+)', expand=False)
```

```python
omdb_initframe.insert(omdb_initframe.columns.get_loc('Awards')+1, 'Oscar_winner', Oscar_winner)
```

```python
omdb_initframe.insert(omdb_initframe.columns.get_loc('Awards')+2, 'Award_nomination', Oscar_nominated)
```

```python
omdb_initframe['Oscar_winner'].replace(np.nan, '', inplace = True)
omdb_initframe['Award_nomination'].replace(np.nan, '', inplace = True)
```
