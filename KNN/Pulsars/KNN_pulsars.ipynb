```python
import numpy as np
import pandas as pd
from sklearn import preprocessing, neighbors
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
```

```python
df = pd.read_csv('E:\\greenfox\\Project_phase\\KNN\\Pulsars\pulsar_stars.csv')
```

```python
df.head(50)
```

```python
X = df.drop(['target_class'],1)
```

```python
y = df['target_class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
clf = neighbors.KNeighborsClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
test = [110, 48.350921, 0.4922154, 0.654784, 2.258361, 16.679328, 21.25878, 93.735705]
test = np.reshape(test, (1, -1))
test.shape
```

```python
prediction = clf.predict(test)[0]
```

```python
def is_pulsar(predict):
    if predict == 0:
        return 'Not pulsar'
    return 'Pulsar'

is_pulsar(prediction)
```

```python
knn_cv = neighbors.KNeighborsClassifier()
```

```python
cv_scores = cross_val_score(knn_cv, X, y, cv = 10)
f'cv_scores mean: {np.mean(cv_scores)}'
```

```python
knn2 = neighbors.KNeighborsClassifier()
```

```python
param_grid = {'n_neighbors' : np.arange(3, 22, 2)}
knn_gscv = GridSearchCV(knn2, param_grid, cv = 10)
knn_gscv.fit(X, y)
```

```python
f'Grid search best "k" parameter: {list(knn_gscv.best_params_.values())[0]}'
```

```python
f'Grid search best score: {knn_gscv.best_score_}'
```
